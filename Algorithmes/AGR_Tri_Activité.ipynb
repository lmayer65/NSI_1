```python
from random import randint


counter = 0

## Tri en ordre croissant
def triTableauCroissant(Tab) :
    changed = False
    global counter
    
    for i in range(0,len(Tab) - 1) :
        if Tab[i] > Tab[i + 1] :
            Tab[i],Tab[i + 1] = Tab[i + 1],Tab[i]
            changed = True
            counter += 1
        
        if changed :
            triTableauCroissant(Tab)
       
    return Tab


# Génération d'un tableau de valeurs aléatoires (ici 30)
T = [randint(0,100) for i in range(30)]
print("Tableau initial :",T,"\n")

# Tableau trié
T = triTableauCroissant(T)
print("Tableau final :", T,"en ", counter,  "opérations")
    
```
