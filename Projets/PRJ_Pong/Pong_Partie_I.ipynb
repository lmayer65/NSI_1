## <u>Activité 3</u> : Quitter le programme grâce à un événement

Jusqu'ici, le programme s'arrêtait tout seul au bout de second. Il existe un système d'événements (programmation événementielle) qui permet d'agir sur le programme grâce aux touches du clavier, des clics d'une souris etc.

1/ <b> Tester</b> l'arrêt du programme suivant en cliquant sur la croix de la fenêtre (il n'affiche qu'une fenêtre noire).

```python
# Importation de la librairie pygame
import pygame,sys
from pygame.draw import *
from pygame.display import *

# Variables globales
size = width, height = 500, 500 # Taille de la fenêtre d'affichage
couleurFond = (0, 0, 0) # Fond noir

# Lancement des fonctionnalités de la librairie pygame
pygame.init()

# Affichage d'une fenêtre de dessin de largeur 500 pixels et de hauteur 500 pixels
screen = set_mode(size)

# Remplissage du fond par la couleur blanche
screen.fill(couleurFond)

# Boucle infinie
while True:
    flip() # Activer le rendu graphique
    pygame.time.delay(15) # Une image toutes les 15 ms

    # Test d'arrêt du jeu
    for event in pygame.event.get():
        if event.type == pygame.QUIT:# si clic sur la croix "quitter"
            pygame.display.quit() # Ferme la fenêtre
            sys.exit() # Arrête le programme
```

## <u>Activité 4</u> : Mettre en mouvement la balle

L'objectif ici est de mettre en place le mouvement en pouvant faire varier les attributs de la balle (position, vitesse, couleur, etc.).
On considère qu'à l'instant t, la balle est à la position (x,y) dans un repère orthogonal (O;I;J). A la frame (affichage) suivant, à l'instant t + dt, sa nouvelle position est (x + dx ; y + dy).

<u> Remarque</u> : dx, dy et dt signifient une variation très faible (variation dite "infinitésimale") de x,y et t.

```python
# Importation de la librairie pygame
import pygame,sys
from pygame.draw import *
from pygame.display import *

# Variables globales ##########################################################"
size = width, height = 500, 500 # taille de la fenêtre d'affichage

# Attributs de la balle
dx,dy = 2, 1 # Vecteur de déplacement entre deux frames consécutives
couleurFond = 255, 0, 0  #Rouge
couleurBalle= 0,0,255    #Bleu
x,y = 100,100 # Position initiale de la balle
r = 20 # Rayon de la balle

#lancement des fonctionnalités de la librairie pygame
pygame.init()

# Affichage d'une fenêtre de dessin de largeur 500 pixels et de hauteur 500 pixels
screen = set_mode(size)

# Remplissage du fond par la couleur blanche
screen.fill(couleurFond)

# Boucle infinie
while True:
    
    # Afficher la balle :
    circle(screen, couleurBalle , (x,y), r, 0)# dessin d'un cercle :

    # Déplacement de la balle :
    x = x + dx
    y = y + dy
    
    flip()# Activer le rendu graphique
    
    pygame.time.delay(15)# Pause de 15 ms
    
    # Test d'arrêt du jeu
    for event in pygame.event.get():
        if event.type == pygame.QUIT:# Si clic sur la croix quitter
            pygame.display.quit()    # Ferme la fenêtre
            sys.exit()               # Arrête le programme
```
